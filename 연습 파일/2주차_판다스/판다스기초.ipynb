```python
import pandas as pd
```

# **Data Frame**

```python
#공식문서(도움말)활용
#pd.DataFrame?
#pd.DataFrame(shift+tab+tab)

df = pd.DataFrame(
{"a" : [4, 5, 6],
"b" : [7, 8, 9],
"c" : [10, 11, 12]},
index = [1, 2, 3])

df
```

## **Series**

```python
df["a"]
#"a"칼럼에 있는 4,5,6이 나옴
# Series 형태
# 1차원 구조
```

```python
df[['a']]
# 대괄호를 하나 더 사용-> 데이터프레임 형태
# 2차원 구조
```

### **Subset** (일부값불러오기)

```python
df[df["a"] > 4]
# "a"에서 4이상만 가져오고 싶음. 
```

```python
df[["a","b"]]
#두개 이상 불러오려면 데이터프레임 형태로 가져오기
```

### **Summarize Data** (값의 빈도수 구하기)

```python
df["a"].value_counts()
```

### **Reshaping** (sort_values, drop)

```python
# "a"컬럼을 기준으로 정렬
df["a"].sort_values()
```

```python
# DataFrame 전체에서 "a"값을 기준으로 정렬하기
df.sort_values("a")
```

```python
# 역순으로 정렬하기
df.sort_values("a", ascending=False)
```

```python
# "c"컬럼 drop 하기
df = df.drop(["c"], axis=1)
df
```
