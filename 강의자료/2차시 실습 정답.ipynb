## 1. 아래의 데이터를 "df_book"이라는 이름의 데이터 프레임으로 만들기

```python
import pandas as pd
```

```python
book1=pd.Series({'도서명':'수상한 주인장',
                 '재고수량': 5,'정가': 16000,
                 '저자': '김주현','출판사': '넥서스'})
book2=pd.Series({'도서명': '친절',
                  '재고수량': 15,'정가': 10000,
                  '저자': '스테판 아인혼','출판사': '마고북스'})
book3=pd.Series({'도서명':'웃음',
                 '재고수량': 50,'정가': 11800,
                 '저자': '베르나르 베르베르','출판사': '예제출판'})
book4=pd.Series({'도서명':'생각에 관한 생각',
                 '재고수량': 85,'정가': 22000,
                 '저자': '대니얼 카너먼','출판사': '김영사'})
book5=pd.Series({'도서명':'그대로 두기',
                 '재고수량': 0,'정가': 12000,
                 '저자': '다이애나 애실','출판사': '열린책들'})
book6=pd.Series({'도서명':'개미',
                 '재고수량': 58,'정가': 8000,
                 '저자': '베르나르 베르베르','출판사': '열린책들'})
```

```python
df_book=pd.DataFrame([book1,book2,book3,book4,book5,book6],
                     index=['0','1','2','3','4','5'])
```

```python
df_book
```

## 2. "df_book"을 이용하여 아래의 조건을 만족시키기(결과는 데이터 프레임의 형태)

### 2-1  책 이름만 표시

```python
df_book[['도서명']]
```

### 2-2 "베르나르 베르베르"에 의해 쓰여진 책

```python
df_book[df_book['저자']=='베르나르 베르베르']
```

### 2-3 "열린책들"에서 출판한 책

```python
df_book[df_book['출판사']=='열린책들']
```

### 2-4 2번과 3번을 모두 만족하는 책

```python
df_book[(df_book['출판사']=='열린책들')&(df_book['저자']=='베르나르 베르베르')]
```

### 2-5 첫번째부터 세번째까지의 책 정보

```python
df_book.iloc[0:3,:]
```

### 2-6 재고가 없는 책

```python
df_book[df_book['재고수량']==0]
```

### 2-7 책의 이름, 목록, 가격

```python
df_book[['도서명','재고수량','정가']]
```

### 2-8 첫 번째, 두 번째, 세 번째 책의 이름, 재고, 가격

```python
df_book.iloc[0:3,0:3]
```

## 3. "df_book2"(3-1에서 만들 예정)를 사용하여 아래 문제를 해결

### 3-1 "df_book"을 "df_book2"로 복사

```python
df_book2=df_book
```

### 3-2 "sale_price" 열을 추가(초기 값을 0으로 설정)

```python
df_book2['sale_price']=0
```

```python
df_book2
```

### 3-3 "sale_price"의 각 값을 각 책 가격의 90%로 설정

```python
df_book2['sale_price']=df_book2['정가']*0.9
```

```python
df_book2
```

## 4."df_book3"(4-1에서 만들 예정)을 사용하여 아래 문제를 해결

### 4-1 "df_book2"에 재고가 있는 책으로 "df_book3"를 생성

```python
df_book3=df_book2[df_book2['재고수량']>0]
```

```python
df_book3.reset_index(drop=True,inplace=True)
df_book3
```

### 4-2 열 이름 "재고수량"을 "재고"로 변경

```python
df_book3.rename(columns={'재고수량':'재고'},inplace=True)
```

```python
df_book3
```

### 4-3 모든 책의 수량을 10권 늘리기

```python
df_book3['재고']+=10
```

```python
df_book3
```

### 4-4 책 "수상한 주인장" 삭제

```python
df_book3.drop(0,inplace=True)
```

```python
df_book3
```

### 4-5 열 "sale_price" 삭제

```python
df_book3.drop(['sale_price'],axis=1,inplace=True)
```

```python
df_book3
```
