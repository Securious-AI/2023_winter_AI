#### df.corr()
각 칼럼들에 대한 상관계수

```python
df.corr()
```

corr함수를 통한 상관계수 보충 설명
* 변수와 상관관계: 변수간 흐름이 얼마나 비슷한가를 나타내는 척도. =a가 증가하면 b도 증가하냐/감소하냐
= 증가의 성향이 얼마나 비슷한가


* 1에 가까울 수록 두 개가 비슷하게 증가
* -1에 가까울 수록 하나가 증가할 수록 다른 하나가 감소
* 0에 가까울 수록 두 개의 관계가 없다


* 숫자형 데이터에 대해서만 연산
* 인과관계를 의미할 수도 아닐 수도 있음

```python
df.value_counts()
#카테고리 값이 존재하는 경우 사용하는 것이 좋음
#각 카테고리별 개수 파악
```
